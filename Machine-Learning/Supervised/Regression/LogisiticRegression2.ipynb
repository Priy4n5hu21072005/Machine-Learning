```python
# Getting the datasets from sklearn
from sklearn.datasets import make_classification
X,y=make_classification(
    n_samples=500,
    n_features=2,
    n_informative=2,
    n_redundant=0,
    n_repeated=0,
    n_classes=2,
    random_state=23
)
```

```python
# Classified data in train and test 
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=23)
```

```python
# Train the model 
from sklearn.linear_model import LogisticRegression
model1=LogisticRegression()
model1.fit(X_train,y_train)
```

```python
# checking the iteration and score of model1 
print("Iteration",model1.n_iter_)
print("score",model1.score(X_test,y_test)*100)
```

```python
# New data points unseen 
new_point = [[0.5, -1.2]]
```

```python
# Predict the values 
prediction = model1.predict(new_point)
print("Predicted Class:", prediction[0])
```
