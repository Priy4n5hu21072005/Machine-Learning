```python
# Genrate datasets
import numpy as np
import pandas as pd
np.random.seed(23)
n=200
# features
X1=np.random.rand(n)*20
X2=np.random.rand(n)*5
X3=X1*X2
# target with noise
y = 3*X1 + 2*X2 + np.random.randn(n)*2
# making dataframe
df=pd.DataFrame(
    {
        "Feature1":X1,
        "Feature2":X2,
        "Feature3":X3,
        "Target":y
    }
)
print(df.head())
```

```python
# Missing value ko add karne ke liye 
df.loc[np.random.choice(df.index,20),"Feature1"]=np.nan
```

```python
# Categorical feature 
df["Category"]=np.random.choice(["A","B"],size=n)
```

```python
# Data cleaning:
    # missing values 
df.fillna({"Feature1":df["Feature1"].mean()},inplace=True)
```

```python
# Encoding
from sklearn.preprocessing import OneHotEncoder
df=pd.get_dummies(df,drop_first=True)
```

```python
#  central scalling
from sklearn.preprocessing import StandardScaler
X=df.drop("Target",axis=1)
y=df["Target"]
```

```python
# Split data in train and test
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=23,test_size=0.2)
scaler=StandardScaler()
X_train=scaler.fit_transform(X_train)
X_test=scaler.transform(X_test)
```

```python
from sklearn.linear_model import Ridge, Lasso
from sklearn.metrics import mean_squared_error

# Ridge
ridge = Ridge(alpha=1.0)
ridge.fit(X_train, y_train)
ridge_pred = ridge.predict(X_test)

# Lasso
lasso = Lasso(alpha=0.1)
lasso.fit(X_train, y_train)
lasso_pred = lasso.predict(X_test)

print("Ridge MSE:", mean_squared_error(y_test, ridge_pred))
print("Lasso MSE:", mean_squared_error(y_test, lasso_pred))
print("Ridge Coef:", ridge.coef_)
print("Lasso Coef:", lasso.coef_)
```

```python
# Predict for entire dataset
import plotly.express as px
df["Predicted"] = ridge.predict(X)

fig = px.scatter_3d(
    df,
    x="Feature1",
    y="Feature2",
    z="Target",
    color="Predicted",
    title="3D Scatter with Prediction Coloring"
)

fig.show()
```

```python
fig = px.scatter_3d(
    df,
    x="Feature1",
    y="Feature2",
    z="Feature3",
    color="Target",
    title="3D View with Color as Target"
)

fig.show()
```

```python
Q1 = df["Target"].quantile(0.25)
Q3 = df["Target"].quantile(0.75)
IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

outliers = df[(df["Target"] < lower) | (df["Target"] > upper)]

print("Number of outliers:", len(outliers))
```
