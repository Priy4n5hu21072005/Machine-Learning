```python
import torch
import torch.nn as nn
```

```python
class Model(nn.Module):
  def __init__(self,num_features):
    super().__init__()
    self.networks=nn.Sequential(
        nn.Linear(num_features,3),
        nn.ReLU(),
        nn.Linear(3,1),
        nn.Sigmoid()
    )
  def forward(self,features):
    output=self.networks(features)
    return output
```

```python
# Random Datasets
features=torch.rand(10,5)
# Model
mod=Model(features.shape[1])
# Forward pass
mod(features)
```
